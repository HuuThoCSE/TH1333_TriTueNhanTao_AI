## **P2.1.** Thư viện  **sklearn.tree**

### **P2.1.1.** Hàm **export_graphviz()**
Hàm export_graphviz để tạo ra đồ thị cây quyết định.

#### **P2.1.1.1. Cú pháp:**
*sklearn.tree.export_graphviz(decision_tree, out_file=None, *, max_depth=None, feature_names=None, class_names=None, label='all', filled=False, leaves_parallel=False, impurity=True, node_ids=False, proportion=False, rotate=False, rounded=False, special_characters=False, precision=3, fontname='helvetica')*

#### **P2.1.1.2. Thuộc tính**:
- **max_depth**: Độ sâu tối đa của cây.
feature_names: Danh sách tên của các đặc trưng được sử dụng trong việc xây dựng cây.
- **class_names**: Danh sách tên của các lớp dữ liệu.
- **filled**: Chỉ định liệu các nút của cây có nền màu hay không.
- **rounded**: Chỉ định liệu các nút của cây có cạnh tròn hay không.
- **special_characters**: Chỉ định liệu có hiển thị các ký tự đặc biệt hay không.
- **out_file**: Tên file đầu ra, nếu None, sẽ trả về chuỗi đại diện cho đồ thị.
- **impurity**: Chỉ định liệu sử dụng chỉ số độ đo đa dạng hay là độ đo độ tinh khiết.
- **proportion**: Chỉ định liệu nên hiển thị phần trăm mẫu của từng lớp dữ liệu.
- **rotate**: Chỉ định liệu có xoay các nhãn của nút hay không.
- **label**: Chỉ định nhãn cho từng nút của cây.
- **node_ids**: Chỉ định liệu có hiển thị ID của nút hay không.

## **P2.2** Thư viện **sklearn.model_selection**

### **P2.2.1** Hàm **train_test_split**
Hàm train_test_split dùng để chia tập dữ liệu thành tập huấn luyện và tập kiểm tra.

#### **P2.2.1.1 Cú pháp:**
*sklearn.model_selection.train_test_split(*arrays, test_size=None, train_size=None, random_state=None, shuffle=True, stratify=None)*

#### **P2.2.1.2 Thông số:**
- **array**: Các mảng dữ liệu cần chia (ví dụ: các mảng đặc trưng, mảng nhãn).
- **test_size**: Các mảng dữ liệu cần chia (ví dụ: các mảng đặc trưng, mảng nhãn).
- **train_size**: Tỉ lệ phần trăm của tập huấn luyện (nếu không được chỉ định thì sẽ lấy phần còn lại của test_size để tạo thành tập huấn luyện).
- **random_state**: Số nguyên để đảm bảo rằng kết quả chia sẽ giống nhau mỗi lần chạy thuật toán (mặc định là **'None'**).
- **shuffle**: Chọn xem liệu dữ liệu sẽ được trộn trước khi chia hay không (mặc định là **True**).
- **stratify**: Dữ liệu nhãn được sử dụng để tạo các phân lớp đối với tập dữ liệu, giúp đảm bảo rằng tỷ lệ phân phối của các lớp giữa các tập huấn luyện và kiểm tra là cân bằng (mặc định là **None**).

* update=False,  # update all models

* line_thickness=3,  # bounding box thickness (pixels)
* hide_labels=False,  # hide labels
* hide_conf=False,  # hide confidences
* half=False,  # use FP16 half-precision inference

* vid_stride=1,  # video frame-rate stride
):


---


* weights: đường dẫn đến trọng số của mô hình YOLOv5
* source: đường dẫn đến tệp ảnh hoặc video hoặc webcam
* data: đường dẫn đến tệp dữ liệu của tập dữ liệu COCO
* imgsz: kích thước ảnh đầu vào cho quá trình suy luận
* **conf_thres**: ngưỡng tin cậy để phát hiện đối tượng # THI!!! | Chỉnh định: chỉ định ngưỡng **tối thiểu**
* **iou_thres**: ngưỡng chồng chéo giữa các hộp giới hạn để áp dụng thuật toán NMS | Chỉnh định: chỉ định ngưỡng **tối đa**
* max_det: số lượng đối tượng tối đa được phát hiện trên mỗi ảnh
* device: thiết bị sử dụng cho quá trình suy luận, có thể là GPU hoặc CPU
* view_img: hiển thị kết quả trên ảnh
* save_txt: lưu kết quả dưới dạng file văn bản
* save_conf: lưu giá trị độ tin cậy của các đối tượng được phát hiện
* save_crop: lưu hình ảnh đã được cắt để phát hiện đối tượng
* nosave: không lưu ảnh hoặc video
* classes: chỉ phát hiện các lớp cụ thể, có thể chỉ định số hoặc tên lớp
* agnostic_nms: áp dụng thuật toán NMS không phụ thuộc vào lớp
* augment: sử dụng kỹ thuật tăng cường dữ liệu để phát hiện đối tượng
* visualize: hiển thị kết quả trên đối tượng
* update: cập nhật tất cả các mô hình
* project: đường dẫn đến thư mục để lưu kết quả
* name: tên thư mục để lưu kết quả
* exist_ok: cho phép ghi đè kết quả cũ nếu đã tồn tại
* line_thickness: độ dày của đường viền cho hộp giới hạn
* hide_labels: không hiển thị nhãn của đối tượng
* hide_conf: không hiển thị giá trị độ tin cậy của đối tượng
* half: sử dụng kiểu dữ liệu độ chính xác nửa (FP16) để phát hiện đối tượng
* dnn: sử dụng thư viện OpenCV DNN cho quá trình suy luận
* vid_stride là số lượng khung hình của video sẽ được bỏ qua trước khi thực hiện phát hiện đối tượng. Ví dụ: nếu vid_stride=2, mỗi khung hình thứ 2 sẽ được bỏ qua để giảm độ phức tạp tính toán.
