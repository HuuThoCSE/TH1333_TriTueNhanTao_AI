**1. Import các thư viện cần thiết:**

cv2: Thư viện OpenCV dùng để triển khai thuật toán KNN.

numpy: Thư viện hỗ trợ các tính toán trên ma trận.

matplotlib: Thư viện hỗ trợ vẽ đồ thị.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

**2. Tạo dữ liệu huấn luyện và nhãn:**

Tạo một ma trận trainData kích thước (25, 2) chứa các giá trị ngẫu nhiên trong khoảng từ 0 đến 100.

Tạo một vector ketqua kích thước (25, 1) chứa các giá trị nhãn ngẫu nhiên bằng 0 hoặc 1.

```python
trainData = np.random.randint(0, 100, (25, 2)).astype(np.float32)
ketqua = np.random.randint(0, 2, (25, 1)).astype(np.float32)
```

**3. Tách dữ liệu thành 2 nhóm dựa trên nhãn:**

Tạo hai ma trận red và blue tương ứng với các mẫu dữ liệu có nhãn là 1 và 0.

```python
red = trainData[ketqua.ravel()==1]
blue = trainData[ketqua.ravel()==0]
```

**4. Tạo một mẫu dữ liệu mới:**

Tạo một ma trận newMember kích thước (1, 2) chứa các giá trị ngẫu nhiên trong khoảng từ 0 đến 100.

```python
newMember = np.random.randint(0, 100, (1, 2)).astype(np.float32)
```

**5. Hiển thị dữ liệu huấn luyện và mẫu dữ liệu mới trên đồ thị Scatter plot:**

Sử dụng hàm plt.scatter để vẽ các điểm dữ liệu lên đồ thị. Các điểm dữ liệu được chia thành hai màu (đỏ và xanh) tương ứng với các nhóm dữ liệu khác nhau. Mẫu dữ liệu mới được biểu diễn bằng màu xanh lá cây.

```python
plt.scatter(red[:,0], red[:,1], 100, 'r', 's')
plt.scatter(blue[:,0], blue[:,1], 100, 'b', '^')
plt.scatter(newMember[:,0], newMember[:,1], 100, 'g', 'o')
```

**6. Khởi tạo mô hình KNN và huấn luyện mô hình:**

Sử dụng lớp cv2.ml.KNearest_create() để khởi tạo một đối tượng KNN.

Sử dụng hàm knn.train để huấn luyện mô hình KNN với dữ liệu huấn luyện và nhãn tương ứng.

```python
knn = cv2.ml.KNearest_create()
knn.train(trainData, 0, ketqua)
```

**7. Dự đoán nhãn của mẫu dữ liệu mới:**

Sử dụng hàm knn.findNearest để dự đoán nhãn của mẫu dữ liệu mới. Tham số đầu vào của hàm này là newMember (mẫu dữ liệu mới), 3 (số lượng hàng xóm gần nhất cần tìm để đưa ra quyết định). Kết quả trả về bao gồm nhãn của mẫu dữ

```python
temp, ketqua, hangxom, khoangcach = knn.findNearest(newMember, 3)
```

```python
print("Result: {}\n".format(ketqua))
print("Hang xom: {}\n".format(hangxom))
print("Khoang cach: {}\n".format(khoangcach))

plt.show()
```
