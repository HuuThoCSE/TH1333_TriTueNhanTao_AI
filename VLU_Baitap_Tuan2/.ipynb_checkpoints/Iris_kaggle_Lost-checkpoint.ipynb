```python
# B1: Khởi tạo trọng số
import random
W_real = []
for i in range(4):
    W_real.append(random.randint(0,1))
print(W_real)
```

```python
df.Species
```

```python
dataset = df.to_numpy()
dataset
```

```python
W_train=[]
theta = 0
anpha = 0.05
for y in range(0, len(dataset)-1):
    print(dataset[y,0])
    W_real = [1,0,0,0]
    W = W_real
    run=True
    kq_real = dataset[y,len(dataset)]
    while(run):
        X = 0
        kq = 0
        for i in range(1, len(dataset)-1):
            X = round(X+dataset[y,i]*W[i-1])
        if X >= theta:
            kq = 1
        elif X < theta:
            kq = -1
        if str(kq) == kq_real:
            W_train.append(W[y])
            print("Đoán đúng. \tX=",X,"\t W",[y+1],"=",W[y], "\t kq=", kq, "\t kq_real= ", kq_real, '\n')
            run = False
        else:
            print("Chưa đoán đúng. X=", X,"\t W",[y+1],"=",W[y], "\t kq=", kq, "\t kq_real= ", kq_real)
            i=0
            for i in range(len(dataset)-1):
                if str(kq) > kq_real:
                    W[i] = round(W[i]-anpha*dataset[0,i+1],2)
                else:
                    W[i] = round(W[i]+anpha*dataset[0,i+1])
```

```python
W_train
```

```python
for i in range(1, 5):
    X = X+dataset[5,i]*W_train[i-1]
    if X >= theta:
        kq = 1
    elif X < theta:
        kq = -1
print(kq)
```
