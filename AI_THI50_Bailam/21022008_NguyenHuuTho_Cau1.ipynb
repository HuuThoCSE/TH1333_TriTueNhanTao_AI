```python
import cv2 
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("/content/drive/MyDrive/AI_THI/de2_c1.png")
cv2_imshow(img)
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
# Sobel Y
kernel_sobelY = np.array([[-1,-2,-1], [0,0,0], [1,2,1]])
img_sobelY = cv2.filter2D(gray, -1, kernel_sobelY)
cv2_imshow(img_sobelY)
```
