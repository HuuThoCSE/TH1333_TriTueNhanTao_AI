```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import graphviz
```

```python
# Đọc dữ liệu từ file CSV vào DataFrame
data = pd.read_csv('/content/drive/MyDrive/AI_1333/AI_Tuan5_2/dataset.csv')
data
```

```python
X = data.iloc[:, 1:-1]  # các cột đầu tiên là các feature
X
```

```python
y = data.iloc[:, -1]  # cột cuối là nhãn
y
```

```python
from sklearn.preprocessing import LabelEncoder 
labelencoder_X = LabelEncoder()
X = X.apply(LabelEncoder().fit_transform)
X
```

```python
# Tạo cây quyết định với thuật toán ID3
dt = DecisionTreeClassifier(criterion='entropy')
dt.fit(X, y)
```

```python
# Test dữ liệu
from sklearn.metrics import accuracy_score
import numpy as np

dt.fit(X.iloc[:,1:5],y) # hàm .fit là Huấn luyện mô hình (train module)
X_in = np.array([0, 0, 0])
y_pred = dt.predict([X_in]) # .predict : Hàm dự đoán . Đưa ra kết quả kiểu thử
y_pred # Trả về nhánh bất kỳ
```

```python
# Đánh giá độ chính xác (độ hiệu quả)
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score

X_train, X_test, y_train, y_test = train_test_split(X.iloc[:,1:], y, test_size=0.2, random_state=42) # iloc[:, 1:] truy cập toàn bộ dòng và toàn bộ cột, đắt đầu từ cột 2 
# dt.fit(X.iloc[:,1:5],y) 
dt1 = DecisionTreeClassifier(criterion='entropy')
dt1.fit(X_train, y_train)
y_pred = dt.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
# chỉ số đánh giá độ chính xác của mô hình phân loại.
print('Accuracy:', accuracy) # Sự chính xác
print('Precision:', precision) # Nó xác định tần suất mà một mô hình là đúng khi dự đoán lớp positive. Precision càng cao, tức là số điểm mô hình dự đoán là positive đều là positive càng nhiều
print('Recall:', recall) # Nó trả lời câu hỏi: Trong số tất cả các nhãn positive có thể có, mô hình xác định đúng bao nhiêu nhãn? Recall được tính bằng công thức: Recall = True Positive / (True Positive + False Negative)
print('F1-score:', f1)
```
