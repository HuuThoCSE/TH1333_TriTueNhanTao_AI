```python
import cv2 
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/drive/MyDrive/AI_Tailieu/src/TCA_tiger.jpg')
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
kernel_edge_detection = np.array([[-1,-1,-1], [-1,8,-1], [-1,-1,-1]])
img_kernel_edge_detection = cv2.filter2D(img, -1, kernel_edge_detection)
cv2_imshow(img_kernel_edge_detection)
```

```python
kernel_sobelX = np.array([[-1,0,1], [-2,0,2], [-1,0,1]]) * 5
img_sobelX = cv2.filter2D(gray, -1, kernel_sobelX)
cv2_imshow(img_sobelX)
```
