```python
from sklearn.datasets import load_iris
from sklearn import tree
from matplotlib import pyplot as plt
```

```python
dataset = load_iris()
x = dataset.data
y = dataset.target
```

```python
classifier = tree.DecisionTreeClassifier(criterion="entropy", max_depth=4, min_samples_leaf=10) # criterion = entropy/gini ; max_depth = 1, 2, ...
classifier.fit(x, y)
fig, ax = plt.subplots(figsize=(10, 10))
tree.plot_tree(classifier, ax=ax, feature_names=["seatal length", "setal width", "patal length", "petal width"], filled=1)
```
