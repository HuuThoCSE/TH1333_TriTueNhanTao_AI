```python
import cv2 
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/drive/MyDrive/AI_1333/AI_Tuan5_2/Hinh.jpg') # Load hình ảnh
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # Chuyển ảnh thành dạng gray và đưa ảnh về ma trận
rows, cols = img.shape[:2] # Lấy kích thước của ảnh
cv2_imshow(img) # Show hình ảnh
```

```python
# Sharpening kernel - Làm sắc nét
kernel_sharpen_2 = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
output5=cv2.filter2D(img,-1,kernel_sharpen_2)
cv2_imshow(output5)
```

```python
kernel_edge_detection = np.array([[-1,-1,-1], [-1,8,-1], [-1,-1,-1]])
img_kernel_edge_detection = cv2.filter2D(img, -1, kernel_edge_detection)
cv2_imshow(img_kernel_edge_detection)
```

```python
img2 = cv2.imread('/content/drive/MyDrive/AI_1333/AI_Tuan5_2/Hinh2.png') # Load hình ảnh
rows, cols = img2.shape[:2] # Lấy kích thước của ảnh
cv2_imshow(img2) # Show hình ảnh
```

```python
median2 = cv2.medianBlur(src=img2, ksize=7)
cv2_imshow(median2)
```

```python
gray2 = cv2.cvtColor(median2, cv2.COLOR_BGR2GRAY) # Chuyển ảnh thành dạng gray và đưa ảnh về ma trận - Nếu lỗi ở đây coi định dạng hình

sobelX = np.array([[-1,0,1], [-2,0,2], [-1,0,1]])
img_sobel_x = cv2.filter2D(gray2, -1, sobelX)
cv2_imshow(img_sobel_x)
```
